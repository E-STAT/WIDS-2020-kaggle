Let's do Simple lightgbm modelling

```python
import lightgbm as lgb
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import auc, accuracy_score, roc_auc_score, roc_curve, f1_score
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, KFold
```

```python
y = train['hospital_death']
X = train.drop(['hospital_death'], axis=1)
```

```python
skf = StratifiedKFold(n_splits = 5)
skf.get_n_splits(X, y)
print(skf)
```

```python
for train_index, test_index in skf.split(X, y):
    print('Train:', train_index, 'Test:', test_index)
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
```

```python
estimator = lgb.LGBMClassifier(learning_rate = 0.15, metric = 'l1', 
                        n_estimators = 24, num_leaves = 38)


param_grid = {
    'n_estimators': [x for x in range(24,40,2)],
    'learning_rate': [0.10, 0.125, 0.15, 0.175, 0.2]}
gridsearch = GridSearchCV(estimator, param_grid)

gridsearch.fit(X_train, y_train,
        eval_set = [(X_test, y_test)],
        eval_metric = ['auc', 'binary_logloss'],
        early_stopping_rounds = 5,
              verbose = 1)
```

```python
print('Best parameters found by grid search are:', gridsearch.best_params_)
```

```python
gbm = lgb.LGBMClassifier(learning_rate = 0.1, metric = 'l1', 
                        n_estimators = 34)


gbm.fit(X_train, y_train,
        eval_set=[(X_test, y_test)],
        eval_metric=['auc', 'binary_logloss'],
early_stopping_rounds=5)
```

```python
test.head()
```

```python
y_pred = gbm.predict_proba((test.drop('hospital_death', axis = 1)))[:, 1]
```

```python
yy = gbm.predict_proba(X_test)[:, 1]
```

```python
roc_auc_score(y_test, yy)
```
